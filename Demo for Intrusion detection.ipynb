# Pose Detection

```python
import cv2
import mediapipe as mp

mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

cap = cv2.VideoCapture(0)

# Set desired frame width and height
frame_width = 1280
frame_height = 720
cap.set(cv2.CAP_PROP_FRAME_WIDTH, frame_width)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, frame_height)

if not cap.isOpened():
    print("Error: Could not open camera.")
    exit()

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()

        if not ret:
            print("Error: Could not read frame from camera.")
            break
        
        # Recolor image to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
      
        # Make detection
        results = pose.process(image)
    
        # Recolor back to BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # Render pose landmarks
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                  mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2), 
                                  mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
                                 )               
        
        cv2.imshow('Camera Feed', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp

mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

cap = cv2.VideoCapture(0)

# Set desired frame width and height
frame_width = 1280
frame_height = 720
cap.set(cv2.CAP_PROP_FRAME_WIDTH, frame_width)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, frame_height)

if not cap.isOpened():
    print("Error: Could not open camera.")
    exit()

selected_points = []   # Keeps track of the selected points for the frame
overlap = False        # Flag to track overlap between frames

def on_click(event, x, y, flags, param):
    global selected_points
    if event == cv2.EVENT_LBUTTONDOWN:
        selected_points.append((x, y))
        if len(selected_points) == 4:
            cv2.setMouseCallback('Mediapipe Feed', lambda *args: None)  # Disable further clicks

cv2.namedWindow('Mediapipe Feed')
cv2.setMouseCallback('Mediapipe Feed', on_click)

def check_overlap(frame_x, frame_y, frame_width, frame_height, green_x, green_y, green_size):
    if (frame_x < green_x + green_size and
        frame_x + frame_width > green_x and
        frame_y < green_y + green_size and
        frame_y + frame_height > green_y):
        return True
    return False

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()

        if not ret:
            print("Error: Could not read frame from camera.")
            break
        
        # Draw the selected points
        for point in selected_points:
            cv2.circle(frame, point, 5, (0, 0, 255), -1)

        # If all four points are selected, create the frame
        if len(selected_points) == 4:
            top_left = selected_points[0]
            top_right = selected_points[1]
            bottom_right = selected_points[2]
            bottom_left = selected_points[3]

            cv2.line(frame, top_left, top_right, (0, 255, 255), 2)
            cv2.line(frame, top_right, bottom_right, (0, 255, 255), 2)
            cv2.line(frame, bottom_right, bottom_left, (0, 255, 255), 2)
            cv2.line(frame, bottom_left, top_left, (0, 255, 255), 2)

        # Recolor image to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
      
        # Make detection
        results = pose.process(image)
    
        # Recolor back to BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # Extract landmarks
        try:
            landmarks = results.pose_landmarks.landmark
            min_x = min(landmark.x for landmark in landmarks)
            max_x = max(landmark.x for landmark in landmarks)
            min_y = min(landmark.y for landmark in landmarks)
            max_y = max(landmark.y for landmark in landmarks)

            # Draw bounding box
            cv2.rectangle(image, (int(min_x * image.shape[1]), int(min_y * image.shape[0])),
                          (int(max_x * image.shape[1]), int(max_y * image.shape[0])),
                          (0, 255, 0), 2)
            
            # Calculate the coordinates for the green pose frame
            green_x = int(min_x * image.shape[1])
            green_y = int(min_y * image.shape[0])
            green_size = int((max_x - min_x) * image.shape[1])
            
            # Check if the yellow frame overlaps with the green frame
            if check_overlap(top_left[0], top_left[1], top_right[0] - top_left[0], bottom_left[1] - top_left[1],
                             green_x, green_y, green_size):
                overlap = True
            else:
                overlap = False
        except:
            pass
        
        # Render detections
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2), 
                                mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
                                 )
        
        # Calculate the coordinates for the center of the video feed
        center_x = image.shape[1] // 2
        center_y = image.shape[0] // 2
        
        # Draw the fixed yellow frame and handle color change if overlapping
        if len(selected_points) == 4:
            frame_x = top_left[0]
            frame_y = top_left[1]
            frame_width = top_right[0] - top_left[0]
            frame_height = bottom_left[1] - top_left[1]
            
            if overlap:
                color = (0, 0, 255)  # Red color if overlapping
            else:
                color = (0, 255, 255)  # Yellow color if not overlapping
            
            cv2.rectangle(image, (frame_x, frame_y),
                          (frame_x + frame_width, frame_y + frame_height),
                          color, 2)
        
        cv2.imshow('Mediapipe Feed', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
```
